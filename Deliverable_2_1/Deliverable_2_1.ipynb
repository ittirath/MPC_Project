## Deliverable 2.1 — Why the Linearized Rocket System Splits Into Four Independent Subsystems

Around the hover equilibrium, the rocket behaves like a *weakly coupled*, almost *block-diagonal* mechanical system. This occurs because each actuator influences a distinct physical direction of motion, and the rocket’s symmetry causes most cross-couplings to vanish when linearized.

### 1. Thrust geometry creates orthogonal force channels

The thrust direction in body coordinates is
$ {}_{b}e_{F}(\delta_1,\delta_2) = $
$
\begin{bmatrix}
\sin\delta_2 \\
-\sin\delta_1\cos\delta_2 \\
\cos\delta_1\cos\delta_2
\end{bmatrix}.
$

For small angles (near hover),


$ {}_{b}e_{F}(\delta_1,\delta_2) \approx $
$
\begin{bmatrix}
\delta_2\\
-\delta_1\\
1
\end{bmatrix}.
$

This means:
- **δ₂** produces horizontal force mainly in **x**,
- **δ₁** produces horizontal force mainly in **y**,
- **Pavg** changes only vertical thrust,
- **Pdiff** creates a pure torque about the body-z axis.

Thus each input primarily excites *one dominant physical direction* of motion.

---

### 2. The rocket is symmetric about its vertical axis

The inertia matrix is diagonal:

$J = \mathrm{diag}(J_x, J_y, J_z).$

This symmetry implies:
- Tilting thrust in the **x-plane** excites **pitch** (β, ω\_y),
- Tilting thrust in the **y-plane** excites **roll-like motion** about x (α, ω\_x),
- Yaw/roll dynamics (γ, ω\_z) are influenced only by **Pdiff**.

Cross-couplings between these axes are *second-order* and vanish exactly at the equilibrium when linearized.

---

### 3. Zero attitude at trim removes geometric coupling

At the hover equilibrium,

$\alpha = \beta = \gamma = 0, \qquad T_{wb} \approx I.$

Body axes align with world axes.  
Thus:
- Lateral body forces directly become world-frame accelerations,
- No rotation-induced mixing between x, y, and z directions occurs at first order.

---

### 4. Each actuator drives one dominant feedback loop

Combining the above, we obtain four nearly independent subsystems:

| Subsystem | Input | States | Physical Meaning |
|----------|--------|--------|------------------|
| sysₓ | δ₂ | (ω\_y, β, v\_x, x) | Tilt in x-plane → pitch → x-acceleration |
| sysᵧ | δ₁ | (ω\_x, α, v\_y, y) | Tilt in y-plane → roll-like motion → y-acceleration |
| sys\_z | Pavg | (v\_z, z) | Vertical thrust only |
| sys\_roll | Pdiff | (ω\_z, γ) | Differential torque → yaw/roll rotation |

In the linearized model, the dynamics of the four motion channels (x, y, z, and roll) become decoupled at first order because all cross-coupling terms vanish at the hover equilibrium. As a result, the A matrix becomes block-diagonal after reordering the states, while the B matrix exhibits a block-column structure, where each input affects only the states of its corresponding subsystem.

---

### 5. Summary

At hover, the rocket’s thrust vectoring geometry, its axial symmetry, and the zero nominal attitude combine to make the linearized dynamics naturally decouple into four independent channels. Each control input impacts only one physical mode at first order, and the spacecraft’s symmetry eliminates the remaining couplings. As a result, the linearized model splits cleanly into the four SISO subsystems (x, y, z, roll), each of which can be controlled independently.
